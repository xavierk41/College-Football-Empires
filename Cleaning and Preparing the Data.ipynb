## 2019-2020 Season Schedule

We need to remove the team ranking in parentheses from the winner and loser columns using regular expressions

```python
# We need to remove the team ranking in parentheses from the winner and loser columns using regular expressions
schedule_data = pd.read_html('https://www.sports-reference.com/cfb/years/2019-schedule.html')
len(schedule_data)
schedule = schedule_data[0]
schedule
```

Function that removes the ranking from the team name in the schedule data.
For the future, we could separate the ranking from the winning and losing team name and add them as separate columns

```python
def preprocess(s):
    re_punc = re.compile(r'\W+')
    re_num = re.compile(r'\d+')
    s = re.sub(re_punc,' ',s)
    s = re.sub(re_num,' ',s)
    s = s.strip()
    return s
```

Formatting the data correctly

```python
winner_vals = []
loser_vals = []
for index, row in schedule.iterrows():
    winner_vals.append(preprocess(row['Winner']))
    loser_vals.append(preprocess(row['Loser']))
schedule['Winner'] = winner_vals
schedule['Loser'] = loser_vals
schedule = schedule[schedule['Rk'] != 'Rk']
# schedule
```

```python
# Convert schedule columns Rk, Wk, Pts, and Pts.1 to numeric
schedule['Rk'] = pd.to_numeric(schedule['Rk'], errors='coerce')
schedule['Wk'] = pd.to_numeric(schedule['Wk'], errors='coerce')
schedule['Pts'] = pd.to_numeric(schedule['Pts'], errors='coerce')
schedule['Pts.1'] = pd.to_numeric(schedule['Pts.1'], errors='coerce')
```

Correcting the final weeks of the season so that every bowl game is the same week of the season, with the exception of the National Championship Game.

```python
# If week is > 17 and < 22, week = 18
# schedule[schedule['Wk'] > 17]
for weekname in range(len(schedule['Wk'])):
    if schedule['Wk'].iloc[weekname] > 17 and schedule['Wk'].iloc[weekname] < 22:
        schedule['Wk'].iloc[weekname] = 18
    elif schedule['Wk'].iloc[weekname] == 22:
        schedule['Wk'].iloc[weekname] = 19
schedule
```

# Calculate the closest fbs school to each county

## Function that takes 4 geographic coordinates and outputs the distance in kilometers

```python
R = 6373.0

def calc_distance(lat1, lon1, lat2, lon2):
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    
    a = (sin(dlat / 2))**2 + cos(lat1) * cos(lat2) * (sin(dlon / 2))**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    distance = R * c
    return distance
```

Now we will find the employ this function to assign each county to a university

```python
closest_team = []
for territory in range(len(counties['County [2]'])):
    dist = 0
    dist_from_uni = 0
    for university in range(len(fbs_teams['team'])):
        dist = calc_distance(counties['Latitude'][territory], counties['Longitude'][territory], fbs_teams['latitude'][university], fbs_teams['longitude'][university])
        if dist_from_uni == 0 or dist < dist_from_uni:
            dist_from_uni = dist
            closest_uni = fbs_teams['team'][university]
    closest_team.append(closest_uni)
```

We need the original team for each county, as well as the stadium coordinates, attached to each county.

```python
counties['Original Empire'] = closest_team
stadium_lat = []
stadium_long = []
for territory in range(len(counties['County [2]'])):
    for university in range(len(fbs_teams['team'])):
        if counties['Original Empire'][territory] == fbs_teams['team'][university]:
            current_lat = fbs_teams['latitude'][university]
            current_long = fbs_teams['longitude'][university]
    stadium_lat.append(current_lat)
    stadium_long.append(current_long)

counties['Original Empire Lat'] = stadium_lat
counties['Original Empire Long'] = stadium_long
```
